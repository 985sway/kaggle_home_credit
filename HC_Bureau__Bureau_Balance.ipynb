```python
from IPython.core.display import display, HTML
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
import helper_functions as hf
from time import time
warnings.filterwarnings('ignore')
```

```python
bb = pd.read_csv('bureau_balance.csv').sort_values(['SK_ID_BUREAU','MONTHS_BALANCE'])
bb = pd.get_dummies(bb)
bb_aggs = bb.groupby('SK_ID_BUREAU').mean()
bb_aggs.head()
```

```python
bb_slopes = hf.get_slopes(bb,bb['SK_ID_BUREAU'].unique(),'SK_ID_BUREAU','MONTHS_BALANCE',bb.columns[2:])
bb_slopes.to_csv('bb_slopes.csv',columns = bb_slopes.columns)
bb_slopes = pd.read_csv('bb_slopes.csv')
bb_slopes = bb_slopes.set_index('SK_ID_BUREAU')
bb_slopes.head()
```

```python
bb_weighted = hf.get_weighted(bb,'SK_ID_BUREAU','MONTHS_BALANCE',bb.columns[2:])
bb_weighted.to_csv('bb_weighted.csv',columns = bb_weighted.columns)#, index = bb_weighted.index)
bb_weighted = pd.read_csv('bb_weighted.csv')
bb_weighted = bb_weighted.set_index('SK_ID_BUREAU')
bb_weighted.head()
```

```python
bb = pd.concat([bb_aggs,bb_slopes,bb_weighted],axis=1)
bb.head()
```

```python
bb.to_csv('bb.csv',columns = bb.columns)
bb = pd.read_csv('bb.csv')
bb.head()
```

```python
bur = pd.read_csv('bureau.csv')

bur['bur_f1'] = bur['DAYS_CREDIT_ENDDATE']-bur['DAYS_ENDDATE_FACT']
bur['bur_f2'] = bur['DAYS_CREDIT']-bur['DAYS_CREDIT_ENDDATE']
bur['bur_f3'] = bur['DAYS_CREDIT']-bur['CREDIT_DAY_OVERDUE']
bur['bur_f4'] = bur['AMT_CREDIT_MAX_OVERDUE']/bur['AMT_CREDIT_SUM']
bur['bur_f5'] = bur['AMT_CREDIT_SUM_DEBT']/bur['AMT_CREDIT_SUM_LIMIT']
bur['bur_f6'] = bur['DAYS_CREDIT_UPDATE']-bur['DAYS_CREDIT']
bur['bur_f7'] = bur['AMT_CREDIT_SUM']/bur['AMT_CREDIT_SUM_LIMIT']
    
bur = bur.merge(bb,on='SK_ID_BUREAU',how='left').sort_values(['SK_ID_CURR','DAYS_CREDIT'])

bur_cats = ['CREDIT_ACTIVE','CREDIT_CURRENCY','CREDIT_TYPE']
bur['CREDIT_ACTIVE'] = hf.reduce_cat(bur,'CREDIT_ACTIVE',2,'OTHER')
bur['CREDIT_CURRENCY'] = hf.reduce_cat(bur,'CREDIT_CURRENCY',2,'OTHER')
bur['CREDIT_TYPE'] = hf.reduce_cat(bur,'CREDIT_TYPE',5,'OTHER')
bur = pd.get_dummies(bur)
bur.head()
```

```python
bur_aggs_dict = {}
for i in bur.columns[2:]:
    bur_aggs_dict[i] = ['min','max','mean','sum','nunique','count']
bur_aggs = hf.get_aggs_df(bur,'SK_ID_CURR',bur_aggs_dict)
```

```python
bur_slopes = hf.get_slopes(bur,bur['SK_ID_CURR'].unique(),'SK_ID_CURR','DAYS_CREDIT',bur.columns[3:])
bur_slopes.to_csv('bur_slopes.csv',columns = bur_slopes.columns)
bur_slopes = pd.read_csv('bur_slopes.csv')
bur_slopes = bur_slopes.set_index('SK_ID_CURR')
bur_slopes.head()
```

```python
bur_weighted = hf.get_weighted(bur,'SK_ID_CURR','DAYS_CREDIT',bur.columns[2:])
bur_weighted.to_csv('bur_weighted.csv',columns = bur_weighted.columns)
bur_weighted = pd.read_csv('bur_weighted.csv')
bur_weighted = bur_weighted.set_index('SK_ID_CURR')
bur_weighted.head()
```

```python
bur = pd.concat([bur_aggs,bur_slopes,bur_weighted],axis=1)
bur = hf.drop_high_corrs(bur,0.95)
bur.columns = ['bur_'+i for i in bur.columns]
bur.to_csv('bur.csv',columns = bur.columns)
bur = pd.read_csv('bur.csv').set_index('SK_ID_CURR')
bur.head()
```
