```python
from IPython.core.display import display, HTML
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
import helper_functions as hf

warnings.filterwarnings('ignore')
```

```python
installs = pd.read_csv('installments_payments.csv').sort_values(
    ['SK_ID_PREV','NUM_INSTALMENT_NUMBER'])
installs['PAY_DIFF'] = installs['AMT_INSTALMENT'] - installs['AMT_PAYMENT']
installs['DAYS_DIFF'] = installs['DAYS_INSTALMENT'] - installs['DAYS_ENTRY_PAYMENT']
installs['PAY_DIFF2'] = [1 if i >= 0 else 0 for i in installs['PAY_DIFF']]
installs['DAYS_DIFF2'] = [1 if i >= 0 else 0 for i in installs['DAYS_DIFF']]
installs['ins_f1'] = installs['PAY_DIFF']/installs['AMT_PAYMENT']
installs = installs.set_index('SK_ID_PREV')
installs['SK_ID_PREV'] = installs.index
installs.head()
```

```python
installs_prev_aggs_dict = {'NUM_INSTALMENT_VERSION':['nunique','count','var'],
                          'DAYS_INSTALMENT':['max','mean',], 
                          'DAYS_ENTRY_PAYMENT':['max','mean'],
                          'AMT_INSTALMENT':['min','max','mean','var','sum','std'], 
                          'AMT_PAYMENT':['min','max','mean','var','sum','std'], 
                          'PAY_DIFF':['min','max','mean','var','sum','std'], 
                          'DAYS_DIFF':['min','max','mean','var','sum','std'], 
                          'PAY_DIFF2':['min','max','mean','var','sum','std'],
                          'DAYS_DIFF2':['min','max','mean','var','sum','std'],
                          'ins_f1':['min','max','mean','var','sum','std']}

installs_prev_aggs = hf.get_aggs_df(installs,'SK_ID_PREV',installs_prev_aggs_dict)
installs_prev_aggs.head()
```

```python
slopes_cols = installs[installs.columns[1:-1]].drop('NUM_INSTALMENT_NUMBER',axis=1).columns
installs_prev_slopes = hf.get_slopes(installs,installs.index.unique(),
                                    'SK_ID_PREV','NUM_INSTALMENT_NUMBER',slopes_cols)
installs_prev_slopes.to_csv('installs_prev_slopes.csv', columns = installs_prev_slopes.columns)

installs_prev_slopes = pd.read_csv('installs_prev_slopes.csv').set_index('SK_ID_PREV')
installs_prev_slopes.head()
```

```python
installs_prev_weighted = hf.get_weighted(installs,'SK_ID_PREV','NUM_INSTALMENT_VERSION', slopes_cols[3:])
installs_prev_weighted.to_csv('installs_prev_weighted.csv', columns = installs_prev_weighted.columns)
installs_prev_Weighted = pd.read_csv('installs_prev_weighted.csv').set_index('SK_ID_PREV')
installs_prev_weighted.head()
```
